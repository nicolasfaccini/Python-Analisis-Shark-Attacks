```python
# Filtrar el dataset para tomar solo los casos desde 1900 hasta ahora
df_shark_attacks = df_shark_attacks[df_shark_attacks['Year'] >= 1900]

# Rellenar los valores nulos
# Para columnas numéricas, usaremos la mediana
# Para columnas categóricas, usaremos la moda
for column in df_shark_attacks.columns:
    if df_shark_attacks[column].dtype == 'object':
        mode_value = df_shark_attacks[column].mode()[0]
        df_shark_attacks[column].fillna(mode_value, inplace=True)
    else:
        median_value = df_shark_attacks[column].median()
        df_shark_attacks[column].fillna(median_value, inplace=True)

# Verificar si aún hay valores nulos
df_shark_attacks.isnull().sum()
```

```python
# Filtrar el dataset para tomar solo los casos desde 1900 hasta ahora
df_shark_attacks = df_shark_attacks[df_shark_attacks['Year'] >= 1900]

# Eliminar filas duplicadas
df_shark_attacks = df_shark_attacks.drop_duplicates()

# Crear un índice jerárquico basado en 'Year' y 'Country'
df_shark_attacks_indexed = df_shark_attacks.set_index(['Year', 'Country'])

# Mostrar las primeras filas del DataFrame con índice jerárquico
df_shark_attacks_indexed.head()
```

```python
# 2. Apply

# Función para simplificar la columna 'Injury'
def simplify_injury(injury):
    if 'no injury' in injury.lower():
        return 'No Injury'
    else:
        return 'Injury'

# Aplicar la función a la columna 'Injury'
df_shark_attacks_indexed['Injury_Simplified'] = df_shark_attacks_indexed['Injury'].apply(simplify_injury)

# Mostrar las primeras filas del DataFrame con la nueva columna
df_shark_attacks_indexed[['Injury', 'Injury_Simplified']].head()
```

```python
# Función ajustada para simplificar la columna 'Injury'
def simplify_injury_adjusted(injury):
    if pd.isnull(injury):
        return 'Unknown'
    elif 'no injury' in injury.lower():
        return 'No Injury'
    else:
        return 'Injury'

# Aplicar la función ajustada a la columna 'Injury'
df_shark_attacks_indexed['Injury_Simplified'] = df_shark_attacks_indexed['Injury'].apply(simplify_injury_adjusted)

# Mostrar las primeras filas del DataFrame con la nueva columna
df_shark_attacks_indexed[['Injury', 'Injury_Simplified']].head()
```

```python
# 1. Índices jerárquicos
# Crear un índice jerárquico basado en 'Year' y 'Country'
df_shark_attacks_multiindex = df_shark_attacks.set_index(['Year', 'Country'])

# Mostrar las primeras filas del DataFrame con índice jerárquico
df_shark_attacks_multiindex.head()
```

```python
# 2. Apply
# Convertir la columna 'Type' a mayúsculas
df_shark_attacks['Type'] = df_shark_attacks['Type'].apply(lambda x: x.upper())

# Mostrar las primeras filas para verificar los cambios
df_shark_attacks[['Type']].head()
```

```python
# Convertir la columna 'Type' a mayúsculas, manejando posibles valores no-string
df_shark_attacks['Type'] = df_shark_attacks['Type'].apply(lambda x: x.upper() if isinstance(x, str) else x)

# Mostrar las primeras filas para verificar los cambios
df_shark_attacks[['Type']].head()
```

```python
# 3. Melt
# Transformar el DataFrame de un formato ancho a un formato largo usando las columnas 'Year', 'Country' y 'Type'
df_shark_attacks_melted = df_shark_attacks.melt(id_vars=['Year', 'Country'], value_vars=['Type'],
                                               var_name='Attribute', value_name='Value')

# Mostrar las primeras filas del DataFrame transformado
df_shark_attacks_melted.head()
```

```python
# 4. Astype
# Cambiar el tipo de datos de la columna 'Year' a string
df_shark_attacks['Year'] = df_shark_attacks['Year'].astype(str)

# Mostrar las primeras filas para verificar los cambios
df_shark_attacks[['Year']].head()
```
