```python
servicio = build("youtube", "v3", developerKey=API_KEY)

solicitud = servicio.commentThreads().list(
    part="snippet",
    videoId=videoID,
    maxResults=100
)

contador = 0
while solicitud is not None:
    respuesta = solicitud.execute()
    comentarios = respuesta.get('items', [])
    for comentario in comentarios:
        texto = comentario['snippet']['topLevelComment']['snippet']['textDisplay']
        if 'afraid' in texto.lower():
            contador += 1
    solicitud = servicio.commentThreads().list_next(solicitud, respuesta)

print(f"Número de comentarios que contienen la palabra 'afraid': {contador}")
```

```python
servicio = build("youtube", "v3", developerKey=API_KEY)

solicitud = servicio.commentThreads().list(
    part="snippet",
    videoId=videoID,
    maxResults=100
)

contador = 0
while solicitud is not None:
    respuesta = solicitud.execute()
    comentarios = respuesta.get('items', [])
    for comentario in comentarios:
        texto = comentario['snippet']['topLevelComment']['snippet']['textDisplay']
        if 'shark' in texto.lower():
            contador += 1
    solicitud = servicio.commentThreads().list_next(solicitud, respuesta)

print(f"Número de comentarios que contienen la palabra 'Shark': {contador}")
```

```python
import pandas as pd


solicitud = servicio.commentThreads().list(
    part="snippet",
    videoId=videoID,
    maxResults=100
)

datos = []
while solicitud is not None:
    respuesta = solicitud.execute()
    comentarios = respuesta.get('items', [])

    for comentario in comentarios:
        texto = comentario['snippet']['topLevelComment']['snippet']['textDisplay']

        if 'afraid' in texto.lower():
            id_comentario = comentario['snippet']['topLevelComment']['id']
            fecha = comentario['snippet']['topLevelComment']['snippet']['publishedAt']
            autor = comentario['snippet']['topLevelComment']['snippet']['authorDisplayName']


            datos.append({
                'Número de Comentario': id_comentario,
                'Texto de Comentario': texto,
                'Fecha del Comentario': fecha,
                'Autor del Comentario': autor
            })


    solicitud = servicio.commentThreads().list_next(solicitud, respuesta)


df = pd.DataFrame(datos)


print(df)

```
