# Estructurando un proyecto de DS-parte III


```python
df_codificado = pd.get_dummies(df2[['Type', 'Country', 'Area', 'Activity', 'Sex ', 'Species ']], drop_first=True)
X = pd.concat([df2['Year'], df_codificado], axis=1)

df2['Fatal (Y/N)'] = df2['Fatal (Y/N)'].replace('UNKNOWN', 'N').fillna('N')

y = df2['Fatal (Y/N)'].map({'Y': 1, 'N': 0})

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)


modelo = LogisticRegression(max_iter=5000)
modelo.fit(X_entrenamiento, y_entrenamiento)
```

```python
from sklearn.metrics import classification_report



y_predicho = modelo.predict(X_prueba)
print(classification_report(y_prueba, y_predicho))
```

Al analizar los resultados del modelo, se pueden encuentar varios puntos de interés. En cuanto a la precisión, el modelo tiene un rendimiento bastante decente cuando se trata de incidentes "No fatales". Predijo que el 83% de los incidentes serían no fatales. Por otro lado, cuando predijo incidentes "Fatales", acertó en el 58% de los casos, mucho menos.

El recall nos dice que mientras que el modelo es bastante preciso al identificar el 92% de los incidentes no fatales reales, su capacidad para detectar incidentes fatales verdaderos es peor, ya que sólo identifica el 35% de ellos. El F1-score, que combina precisión y recall, nos muestra un equilibrio de 0.87 para los incidentes no fatales y 0.44 para los fatales, sugiriendo que la identificación de situaciones fatales podría mejorarse.

Al descomponer el conjunto de prueba, encontramos que está compuesto por 1254 incidentes no fatales y 369 fatales. El promedio ponderado y el simple nos dan una idea más clara sobre el equilibrio del modelo en función de la cantidad de datos reales disponibles para cada clase.

Con una precisión general del 79%, el modelo parece tener un rendimiento aceptable. Sin embargo, es crucial reconocer sus limitaciones, sobre todo cuando se trata de prever incidentes fatales. Aunque predice incidentes no fatales con eficacia, podría estar pasando por alto muchas situaciones potencialmente peligrosas. En un contexto de seguridad y prevención, este no es un detalle menor, y es fundamental ser consciente de que el modelo podría no ser lo suficientemente robusto para identificar todas las situaciones fatales.

```python
##random forest
from sklearn.ensemble import RandomForestClassifier

df2['Fatal (Y/N)'] = df2['Fatal (Y/N)'].fillna('N')
y = df2['Fatal (Y/N)'].map({'Y': 1, 'N': 0})

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)

modelo_rf = RandomForestClassifier(random_state=42)
modelo_rf.fit(X_entrenamiento, y_entrenamiento)

y_predicho_rf = modelo_rf.predict(X_prueba)
print(classification_report(y_prueba, y_predicho_rf))
```

modelo de Random Forest: en términos de ataques "No Fatales", las cifras se mantienen bastante parejas al modelo anterior. igualmente, el modelo de Random Forest mejora un poco en el recall para ataques "Fatales", 38 contra 35.

ambos modelos pueden prever los casos no fatales, pero les cuesta mas con los fatales.
