```python
import pandas as pd

path = "/content/drive/MyDrive/Data Science CoderHouse/Datasets + IMG + Docu/Shark attacks.csv"
df1 = pd.read_csv(path, encoding='latin1')

df_tiburones = df1
```

```python
# Filtrando el dataset para trabajar solo con los casos de 1900 en adelante, y hacemos un analisis exploratorio inicial para entender las columnas y como esta formado el dataset
df_tiburones = df_tiburones[df_tiburones['Year'] >= 1900]
df_tiburones.head()
```

```python
# Vemos las dimensiones y los datos de las columnas

dimensiones = df_tiburones.shape
tipos_datos = df_tiburones.dtypes

# Vemos las estadisticas basicas de cada columna, para ver que, por ejemplo, los paises unicos son 176. el que mas se repite es USA, lo mismo con Florida, y hay 697 areas unicas.
# Tambien vemos, en sexo, que hay 6 valores unicos, y en Fatal, que hay 7 valores unicos. Esto vamos a tener que trabajarlo ya que claramente ambos casos son un error

estadisticas = df_tiburones.describe(include='all')

dimensiones, tipos_datos, estadisticas.head()
```

```python
# Ahora, buscamos saber cuantos duplicados hay y que nos lo muestre, y vemos que no hay ningun duplicado. Esto es algo raro, pero no encontre otra forma de verlos. Asumo que esta bien.
duplicados = df_tiburones[df_tiburones.duplicated()]


num_duplicados = duplicados.shape[0]
duplicados.head(), num_duplicados
```

```python
# Identificación de columnas con valores nulos y vemos el numero y porcentajes de nulos por columna
nulos_por_columna = df_tiburones.isnull().sum()
porcentaje_nulos = (nulos_por_columna / len(df_tiburones)) * 100

nulos_por_columna, porcentaje_nulos
```

```python
# Para tratar los nulos, hacemos lo que aprendimos multiples veces en el curso:
# Para columnas numéricas, reemplazamos los nulos con la mediana
# Para columnas categóricas, reemplazamos los nulos con la moda

for columna in df_tiburones.select_dtypes(include=['float64', 'int64']).columns:
    mediana = df_tiburones[columna].median()
    df_tiburones[columna].fillna(mediana, inplace=True)


for columna in df_tiburones.select_dtypes(include=['object']).columns:
    moda = df_tiburones[columna].mode()[0]
    df_tiburones[columna].fillna(moda, inplace=True)

# Verificamos si quedaron nulos, vemos que no
df_tiburones.isnull().sum()
```
