#Resumen Ejecutivo


**Conclusiones e insights sobre los datos obtenidos durante todo el proceso**

Entre 1900 y 2017, se registraron ataques de tiburones durante 118 años distintos, con un promedio anual de aproximadamente 45.84 ataques, pero con una alta variabilidad anual. La tendencia muestra un aumento en los ataques en años recientes, posiblemente debido a mejores métodos de registro. Los Estados Unidos lideran significativamente el número de ataques registrados, seguido de Australia y Sudáfrica, lo que podría deberse a diferencias en la población de tiburones, rigor en la documentación o el número de personas que participan en actividades acuáticas.

Al analizar la distribución por edad, las víctimas suelen ser jóvenes, con un promedio de 27 años, aunque hay una amplia variabilidad. Los datos sugieren que los jóvenes podrían estar más expuestos debido a su participación en actividades acuáticas, aunque todos los grupos de edad están en riesgo.

En cuanto a la naturaleza de los ataques, la mayoría no fueron provocados, y hay una relación notable con actividades como el surf y la natación. La edad promedio en incidentes provocados es mayor en comparación con los no provocados, indicando que las actividades que podrían provocar ataques son realizadas más frecuentemente por adultos jóvenes o de mediana edad.

Además, hay un predominio de ataques a hombres (89,5%). Esto podría estar relacionado con las actividades predominantes o roles sociales en actividades marinas.

En Florida, la tendencia de ataques de tiburones muestra patrones similares a los observados globalmente. La incidencia aumentó con el tiempo, posiblemente debido a mejoras en los registros, alcanzando un pico en 2003. La edad promedio de las víctimas es ligeramente menor en Florida (25 años) en comparación con el promedio global (27 años), sugiriendo una participación más alta de jóvenes en actividades acuáticas en esta región.

La mayoría de los ataques en Florida no son fatales, consistentes con la tendencia global, pero enfatizan que los encuentros son principalmente no mortales (923 casos no mortales frente a 53 fatales). Esto refuerza la perspectiva de que, aunque hay un riesgo inherente, es menos probable que resulte en fatalidades.

En cuanto a las actividades, el surf está relacionado con más incidentes en Florida en comparación con las cifras globales, seguido por categorías diversas bajo "Otros", natación y vadeo. Esto podría reflejar la popularidad del surf y otras actividades acuáticas en Florida, requiriendo una mayor conciencia y medidas de seguridad.

Además, al igual que en el escenario global, los ataques en Florida se dirigen desproporcionadamente a los hombres (839 ataques a hombres en comparación con 148 a mujeres), lo que puede estar relacionado con quiénes participan más en actividades acuáticas asociadas con riesgos de tiburones.

En general, los datos de Florida muestran tendencias similares a los hallazgos globales pero destacan una participación más joven y un riesgo elevado asociado con el surf, subrayando la importancia de la educación preventiva y la seguridad en zonas propensas a tiburones.


**Modelos de ML**

En el análisis global, la Regresión Logística y el Random Forest presentaron precisiones del 81% y 82%, respectivamente, para incidentes no fatales. En ataques fatales, la Regresión Logística fue ligeramente superior con un 57% de precisión comparado con el 54% de Random Forest. El recall para incidentes no fatales fue mejor en la Regresión Logística (94%) que en Random Forest (92%), mientras que Random Forest superó en ataques fatales con un 31% contra el 27% de la Regresión Logística. Ambos modelos mostraron un F1-Score similar en ataques no fatales (0.87), pero Random Forest lideró en ataques fatales con 0.39 frente al 0.36 de la Regresión Logística. La exactitud general fue de 79% para la Regresión Logística y 78% para Random Forest.

Centrándonos en Florida, la Regresión Logística experimentó una disminución en su eficacia, con una precisión del 33% en ataques fatales y un F1-Score de 0.11, una caída comparada con su desempeño global. Esta reducción destaca la necesidad de modelos adaptados a las características regionales, considerando que factores locales pueden afectar significativamente los resultados de predicción.

# Estructurando un proyecto de DS-parte III - Regresion Logistica y RandomForest


```python
df_codificado = pd.get_dummies(df2[['Type', 'Country', 'Area', 'Activity', 'Sex ']], drop_first=True)
X = pd.concat([df2['Year'], df_codificado], axis=1)

df2['Fatal (Y/N)'] = df2['Fatal (Y/N)'].replace('UNKNOWN', 'N').fillna('N')

y = df2['Fatal (Y/N)'].map({'Y': 1, 'N': 0})

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)


modelo = LogisticRegression(max_iter=5000)
modelo.fit(X_entrenamiento, y_entrenamiento)
```

```python
num_casos_entrenamiento = X_entrenamiento.shape[0]
print("Número de casos de entrenamiento:", num_casos_entrenamiento)

num_casos_entrenamiento = X_prueba.shape[0]
print("Número de casos de prueba:", num_casos_entrenamiento)
```

```python
y_predicho = modelo.predict(X_prueba)
print(classification_report(y_prueba, y_predicho))
```

Al analizar los resultados del modelo de regresion logistica, se destacan varios puntos clave. En términos de precisión, el modelo demuestra un rendimiento sólido cuando se trata de predecir incidentes "No fatales", con un 81% de precisión. Esto significa que aproximadamente el 81% de los ataques que se clasificaron como no fatales eran, de hecho, no fatales. Sin embargo, el rendimiento en la predicción de incidentes "Fatales" es notablemente inferior, con una precisión del 57%. Esto indica que aproximadamente el 57% de los ataques clasificados como fatales resultaron ser fatales en la realidad.

El recall es una métrica importante para evaluar la capacidad del modelo para detectar casos reales de cada clase. En este sentido, el modelo logra una alta tasa de recall del 94% para incidentes "No fatales". Esto significa que el 94% de los ataques no fatales reales se identifican correctamente como tales. Sin embargo, en el caso de incidentes "Fatales", el recall es sustancialmente más bajo, con solo un 27%. Esto indica que solo se logra detectar el 27% de los ataques fatales reales.

El F1-score combina precisión y recall en una sola métrica, lo que proporciona una evaluación equilibrada del rendimiento del modelo. Para incidentes "No fatales", el F1-score es de 0.87, lo que indica un buen equilibrio entre precisión y recall. En contraste, el F1-score para incidentes "Fatales" es de 0.36, lo que sugiere que la identificación de situaciones fatales puede mejorarse.

El conjunto de prueba se compone de 1,254 incidentes no fatales y 369 fatales, lo que refleja un desequilibrio en la distribución de clases. Por lo tanto, se utilizaron medidas de promedio ponderado y simple para tener en cuenta esta disparidad al evaluar el rendimiento del modelo en función del número de datos reales disponibles para cada clase.

En general, el modelo logra una precisión general del 79%, lo que indica un rendimiento aceptable en la clasificación de incidentes de tiburones. Sin embargo, es esencial reconocer sus limitaciones, especialmente en la predicción de incidentes fatales. Aunque es efectivo en la identificación de ataques no fatales, podría estar pasando por alto muchas situaciones potencialmente peligrosas. Esto es un aspecto crítico en un contexto de seguridad y prevención, y se debe tener en cuenta que el modelo podría no ser lo suficientemente robusto para identificar todas las situaciones fatales. Es importante considerar estas limitaciones al utilizar este modelo en aplicaciones prácticas relacionadas con la seguridad en entornos marinos.



```python
##random forest
from sklearn.ensemble import RandomForestClassifier

df2['Fatal (Y/N)'] = df2['Fatal (Y/N)'].fillna('N')
y = df2['Fatal (Y/N)'].map({'Y': 1, 'N': 0})

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)

modelo_rf = RandomForestClassifier(random_state=42)
modelo_rf.fit(X_entrenamiento, y_entrenamiento)

y_predicho_rf = modelo_rf.predict(X_prueba)
print(classification_report(y_prueba, y_predicho_rf))
```

Las diferencias entre los modelos de Regresión Logística y Random Forest en la clasificación de ataques de tiburones como fatales o no fatales son pocas.

En cuanto a la precisión, la Regresión Logística logra un 81% de precisión para identificar ataques como "No fatales", mientras que Random Forest obtiene un 82%, lo que indica un rendimiento ligeramente superior en esta categoría. Sin embargo, en la clasificación de ataques como "Fatales", la Regresión Logística supera a Random Forest con una precisión del 57% frente al 54% de Random Forest.

En términos de recall, la Regresión Logística muestra un rendimiento mejor en la detección de ataques no fatales reales, con un 94% de recall, en comparación con el 92% de Random Forest. Sin embargo, en la identificación de ataques fatales reales, Random Forest supera a la Regresión Logística con un recall del 31% frente al 27% de la Regresión Logística.

El F1-Score, que combina precisión y recall, es similar para la clasificación de ataques no fatales entre ambos modelos, con un valor de 0.87. Sin embargo, en la clasificación de ataques fatales, Random Forest obtiene un F1-Score ligeramente mejor, con un valor de 0.39, en comparación con el 0.36 de la Regresión Logística.

En cuanto a la exactitud general, la Regresión Logística logra un 79%, mientras que Random Forest alcanza un 78%.

```python
from sklearn.metrics import confusion_matrix


conf_matrix_log = confusion_matrix(y_prueba, y_predicho)
conf_matrix_rf = confusion_matrix(y_prueba, y_predicho_rf)

def plot_confusion_matrix(conf_matrix, model_name):
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues',
                xticklabels=['No Fatal', 'Fatal'], yticklabels=['No Fatal', 'Fatal'])
    plt.xlabel('Predicción')
    plt.ylabel('Real')
    plt.title(f'Matriz de Confusión - {model_name}')
    plt.show()
```

```python
plot_confusion_matrix(conf_matrix_log, 'Regresión Logística')
```

```python
plot_confusion_matrix(conf_matrix_rf, 'Random Forest')
```

Aca vemos las matrices de confusion para ambos modelos.

```python
# Imprimir las matrices de confusión
print("Matriz de Confusión Regresión Logísticna:")
print(conf_matrix_log, "\n")


print("Matriz de Confusión Random Forest:")
print(conf_matrix_rf)
```

En la matriz de confusión de la Regresión Logística, se observa un alto número de verdaderos negativos con 1180 casos, lo que sugiere que el modelo acierta en clasificar la mayoría de los ataques como no fatales cuando realmente son no fatales. Sin embargo, la Regresión Logística tiene 271 falsos negativos (FN), lo que indica que predice incorrectamente algunos ataques fatales como no fatales.

Por otro lado, en la matriz de confusión del Random Forest, se observa un mayor número de verdaderos Positivos (TP) con 114 casos en comparación con la Regresión Logística. Esto sugiere que Random Forest es mejor para identificar ataques fatales reales. Sin embargo, el Random Forest también tiene 99 falsos Positivos (FP), lo que significa que predice incorrectamente algunos ataques no fatales como fatales.
